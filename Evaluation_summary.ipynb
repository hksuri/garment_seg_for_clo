```python
import pandas as pd
import os
os.chdir('/content/drive/MyDrive/CS543-FinalProject')
import utils_deepfashion as utils
```

```python
files= os.listdir('evaluation_csv')
master= pd.DataFrame()
for file_ in files:
  df= pd.read_csv(os.path.join('evaluation_csv', file_))
  df['file_name'] = file_
  for i in range(0,13):
   df.rename(columns= {str(i): utils.deepfashion_dict[i+1]}, inplace=True)
  df.rename(columns = {df.columns[0]: 'metrics'},inplace=True)
  master= pd.concat([df, master])
```

```python
master= master.groupby(['metrics', 'threshold', 'file_name']).mean()
master['mean']= master.loc[:, list(utils.deepfashion_dict.values())].mean(axis=1)
```

```python
master.to_csv('final_eval.csv')
```
